```python
import scipy
import scipy.stats as st
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
normal=st.norm()
normal.cdf(0)
```

```python
normal.sf(0) #1-cdf
```

```python
x=np.arange(-1,1.1,0.1)
plt.plot(x,normal.pdf(x))
```

```python
plt.plot(x,normal.cdf(x))
```
