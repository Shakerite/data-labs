```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
cols = ['mpg','cylinders','displacement','horsepower','weight','acceleration',
        'model_year','origin','car_name']
auto = pd.read_csv(url, sep='\\s+', names=cols)
auto.head()
```

```python
auto.index
```

```python
#changing the index
from pandas import RangeIndex
auto.index=RangeIndex(start=0, stop=398*2, step=2)
auto.head()
```

```python
#changing the index without range Index
auto.index=range(0,398*3,3)
auto.head()
```

```python
auto.index=auto.car_name #index is the car_name
auto.head()
```

```python
auto.drop('car_name',axis=1)#dropping car_name column
```

```python
auto.head()
```

```python
auto.set_index('car_name',inplace=True)#supprime la colonne et en fait l'index
auto.head()
```

```python
(auto.car_name)
```

```python
#having a 2 index dataframe
(auto.car_name+auto.model_year.map(str)).value_counts()
auto.drop_duplicates(subset=['car_name','model_year'],inplace=True)
```

```python
(auto.car_name+auto.model_year.map(str)).value_counts()
auto.set_index(['car_name','model_year'],inplace=True)
auto.head()
```
