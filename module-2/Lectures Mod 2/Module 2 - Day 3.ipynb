```python
#on change la fonction appliquée par défaut
df.pivot_table(index=['Vehicle Class'], columns=['Combined MPG'],aggfunc=np.sum)
```

```python
df.pivot_table(index=['Vehicle Class'], columns=['Cylinders'], values=['Combined MPG'])
```

```python
#Filling the Nan
df.pivot_table(index=['Vehicle Class'], columns=['Cylinders'], values=['Combined MPG'],fill_value=0).apply(round)
```
