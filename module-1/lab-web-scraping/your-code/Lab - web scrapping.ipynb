````python
city=[i.text.strip().split(", ")[0] for i in soup_hack.select('div.row p:has(i.fa-map-marker-alt)')]
country= country=[i.text.strip().split(", ")[1] for i in soup_hack.select('div.row p:has(i.fa-map-marker-alt)')]
city
````
